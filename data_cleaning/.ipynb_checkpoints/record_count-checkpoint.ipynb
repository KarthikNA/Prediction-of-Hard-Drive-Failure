```python
import pandas as pd
import numpy as np
```

```python

from os import listdir
from os.path import isfile, join
```

```python

file_paths = ["../../drive_stats_2019_Q1", "../../data_Q2_2019", "../../data_Q3_2019", "../../data_Q4_2019"]

files = []

for filepath in file_paths:
    files_1 = [f for f in listdir(filepath) if isfile(join(filepath, f))]
    files_1 = [filepath+ "/" + f for f in files_1 if "csv" in f and f.startswith("201")] 
    files.extend(files_1)
len(files)
```

```python
!pwd
```

```python
total = 0
```

```python
# iterate over each file and drop the unwanted columns
for file_name in files:
    data = pd.read_csv(file_name)
    total += data.shape[0]
    print("{} added to master_list::: record_count : {} ::: total : {}".format(file_name, data.shape[0], total))    
```
